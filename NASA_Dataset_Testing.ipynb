```python
import os
import pandas as pd

# Define the root folder path
test_folder_path = '/content/drive/MyDrive/test/data/'

# Example function to read and combine CSV files from subfolders
def read_and_combine_csvs(test_folder_path):
    combined_csv_df = pd.DataFrame()  # Initialize an empty DataFrame

    # Loop through each grade folder
    grade_folders = [f for f in os.listdir(test_folder_path) if os.path.isdir(os.path.join(test_folder_path, f))]
    for grade_folder in grade_folders:
        grade_folder_path = os.path.join(test_folder_path, grade_folder)

        # Loop through files in the grade folder
        for file in os.listdir(grade_folder_path):
            file_path = os.path.join(grade_folder_path, file)

            if file.endswith('.csv'):  # Process only CSV files
                print(f"Attempting to read: {file_path}")

                if os.path.exists(file_path):  # Check if the file exists
                    try:
                        # Read the CSV file into a DataFrame
                        df = pd.read_csv(file_path)
                        combined_csv_df = pd.concat([combined_csv_df, df], ignore_index=True)
                        print(f"Successfully loaded {file_path}")
                    except Exception as e:
                        print(f"Error reading {file_path}: {e}")
                else:
                    print(f"File not found: {file_path}")

    return combined_csv_df

# Run the function
combined_csv_df = read_and_combine_csvs(test_folder_path)

# Display the combined DataFrame (optional)
print(combined_csv_df.head())
```

```python
import os
import numpy as np
from obspy import read

# Define the root folder path
test_folder_path = '/content/drive/MyDrive/test/data/'

# Function to compute additional statistical features
def compute_features(tr):
    # Extract data from trace
    trace_data = tr.data

    # Compute statistical features
    mean_val = np.mean(trace_data)
    max_val = np.max(trace_data)
    min_val = np.min(trace_data)
    std_val = np.std(trace_data)
    rms_val = np.sqrt(np.mean(np.square(trace_data)))
    peak_to_peak_val = max_val - min_val
    energy_val = np.sum(np.square(trace_data))

    # Return as a dictionary of features
    features = {
        'mean': mean_val,
        'max': max_val,
        'min': min_val,
        'std': std_val,
        'rms': rms_val,
        'peak_to_peak': peak_to_peak_val,
        'energy': energy_val
    }

    return features

# Example function to read and combine mseed files with feature extraction
def read_and_combine_mseed_files(test_folder_path):
    combined_mseed_data = []  # Initialize an empty list to hold the mseed features

    # Loop through each grade folder
    grade_folders = [f for f in os.listdir(test_folder_path) if os.path.isdir(os.path.join(test_folder_path, f))]
    for grade_folder in grade_folders:
        grade_folder_path = os.path.join(test_folder_path, grade_folder)

        # Loop through files in the grade folder
        for file in os.listdir(grade_folder_path):
            file_path = os.path.join(grade_folder_path, file)

            if file.endswith('.mseed'):  # Process only .mseed files
                print(f"Attempting to read: {file_path}")

                if os.path.exists(file_path):  # Check if the file exists
                    try:
                        # Read the mseed file using obspy
                        st = read(file_path)

                        # Compute features for each trace in the stream
                        file_features = []
                        for tr in st:
                            features = compute_features(tr)
                            file_features.append(features)

                        # Aggregate features across all traces in the file (optional)
                        # Example: Calculate the mean of all features for the file
                        file_features_mean = {key: np.mean([f[key] for f in file_features]) for key in file_features[0].keys()}

                        # Add the aggregated features (or individual features) to the combined list
                        combined_mseed_data.append(file_features_mean)

                        print(f"Successfully processed {file_path}")
                    except Exception as e:
                        print(f"Error reading {file_path}: {e}")
                else:
                    print(f"File not found: {file_path}")

    # Return the combined features data
    return combined_mseed_data

# Run the function
combined_mseed_data = read_and_combine_mseed_files(test_folder_path)

# If you want to display the resulting feature data
import pandas as pd
mseed_features_df = pd.DataFrame(combined_mseed_data)

# Display the first few rows of the combined DataFrame
print(mseed_features_df.head())
```

```python
import os
import pandas as pd
import numpy as np
from obspy import read

# Path to the 'test' folder on Google Drive (update the folder path accordingly)
test_folder_path = '/content/drive/MyDrive/test/data'

# Initialize an empty list to store the data from all .mseed files
combined_mseed_data = []

# Iterate through each subfolder (S16Grade_A, S16Grade_B, etc.)
grade_folders = [folder for folder in os.listdir(test_folder_path) if os.path.isdir(os.path.join(test_folder_path, folder))]

# Loop through each grade folder
for grade_folder in grade_folders:
    grade_folder_path = os.path.join(test_folder_path, grade_folder)

    # List all mseed files in the current grade folder
    mseed_files = [f for f in os.listdir(grade_folder_path) if f.endswith('.mseed')]

    # Loop through each mseed file
    for mseed_file in mseed_files:
        mseed_file_path = os.path.join(grade_folder_path, mseed_file)

        try:
            # Read the mseed file using ObsPy
            st = read(mseed_file_path)

            # Extract the raw data from the mseed file (assuming you want all traces)
            for tr in st:
                # Convert the trace data to a DataFrame (each mseed file is a Trace object)
                trace_data = pd.DataFrame({'Time': tr.times(), 'Amplitude': tr.data})
                trace_data['File Name'] = mseed_file  # Add a column for the file name
                trace_data['Grade Folder'] = grade_folder  # Add a column for the grade folder

                # Append this data to the combined list
                combined_mseed_data.append(trace_data)

        except Exception as e:
            print(f"An error occurred while processing {mseed_file_path}: {e}")

# Combine all the data from the mseed files into a single DataFrame
combined_mseed_df = pd.concat(combined_mseed_data, ignore_index=True)

# Display the first few rows of the combined mseed data
print(combined_mseed_df.head())
```

```python
import numpy as np
import pandas as pd
from obspy import read
import os

def read_and_flatten_mseed_files(test_folder_path):
    # Initialize an empty list to hold the flattened mseed data and file names
    flattened_mseed_data_list = []
    file_names = []  # To hold the corresponding file names for each mseed data

    # Loop through each folder (S16Grade_A, S16Grade_B, etc.)
    for folder in os.listdir(test_folder_path):
        folder_path = os.path.join(test_folder_path, folder)

        # Check if it's a directory
        if os.path.isdir(folder_path):
            # Loop through each mseed file in the current folder
            for file_name in os.listdir(folder_path):
                if file_name.endswith(".mseed"):
                    file_path = os.path.join(folder_path, file_name)
                    try:
                        # Read the mseed file
                        st = read(file_path)

                        # Flatten each trace data and append to the list
                        flattened_data = np.concatenate([trace.data for trace in st])
                        flattened_mseed_data_list.append(flattened_data)
                        file_names.append(file_name)
                    except Exception as e:
                        print(f"Error reading {file_name}: {e}")

    # Convert the flattened data into a NumPy array
    return flattened_mseed_data_list, file_names

# Path to your test folder
test_folder_path = '/content/drive/MyDrive/test/data'

# Run the function to get the flattened mseed data and corresponding file names
flattened_mseed_data, mseed_file_names = read_and_flatten_mseed_files(test_folder_path)

# If needed, convert it into a DataFrame for further use
flattened_mseed_df = pd.DataFrame({
    'File Name': mseed_file_names,
    'Flattened Data': flattened_mseed_data
})

# Show the DataFrame
print(flattened_mseed_df.head())
```
